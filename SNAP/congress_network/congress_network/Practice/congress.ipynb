```python
import os
import pandas as pd
import numpy as np
import json
import networkx as nx
import matplotlib.pyplot as plt
```

```python
last_dir  =os.path.dirname(os.getcwd())
with open(last_dir+"\\congress_network_data.json") as f:
    json_net = f.read()
net = json.loads(json_net)
net = net[0]
```

```python
index_dict = dict(zip(net["usernameList"], range(len(net["usernameList"]))))
index_dict
```

```python
dg = nx.DiGraph()
dg.add_nodes_from(range(len(net["usernameList"])))

for i, nodes in enumerate(net["inList"]):
    dg.add_edges_from((node, i) for node in nodes)
    
len(dg.nodes), len(dg.edges)
```

```python
plt.figure(figsize=(20, 20))  # 
pos = nx.spring_layout(dg)  # 
nx.draw(dg, pos, with_labels=True, font_size=10)
plt.savefig("congress.png", dpi=300, bbox_inches='tight')
plt.show()
```
